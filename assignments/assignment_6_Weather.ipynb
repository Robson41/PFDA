```python
data_url = "https://cli.fusio.net/cli/climate_data/webdata/hly4935.csv"
df = pd.read_csv(data_url)

df.head()
```

```python
df.columns
```

```python
df['Date'] = pd.to_datetime(df['Date/Time'])
df.head()
```

```python
plt.figure(figsize=(14,5))
plt.plot(df['Date'], df['Air Temperature'])
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.title("Hourly Air Temperature at Knock Airport")
plt.show()
```

```python
df['Day'] = df['Date'].dt.date
daily_mean_temp = df.groupby('Day')['Air Temperature'].mean()

plt.figure(figsize=(14,5))
daily_mean_temp.plot()
plt.xlabel("Day")
plt.ylabel("Mean Temperature (°C)")
plt.title("Daily Mean Temperature at Knock Airport")
plt.show()
```

```python
df['Month'] = df['Date'].dt.to_period('M')
monthly_mean_temp = df.groupby('Month')['Air Temperature'].mean()

plt.figure(figsize=(14,5))
monthly_mean_temp.plot(marker='o')
plt.xlabel("Month")
plt.ylabel("Mean Temperature (°C)")
plt.title("Monthly Mean Temperature at Knock Airport")
plt.show()
```

```python
df['Windspeed'].isna().sum()
```

```python
plt.figure(figsize=(14,5))
plt.plot(df['Date'], df['Windspeed'])
plt.xlabel("Date")
plt.ylabel("Windspeed (knots)")
plt.title("Hourly Windspeed at Knock Airport")
plt.show()
```

```python
rolling_wind = df.set_index('Date')['Windspeed'].rolling(24).mean()

plt.figure(figsize=(14,5))
plt.plot(rolling_wind)
plt.xlabel("Date")
plt.ylabel("Rolling 24h Windspeed")
plt.title("24-Hour Rolling Mean Windspeed")
plt.show()
```

```python
daily_max_wind = df.groupby('Day')['Windspeed'].max()

plt.figure(figsize=(14,5))
daily_max_wind.plot()
plt.xlabel("Day")
plt.ylabel("Max Windspeed")
plt.title("Daily Maximum Windspeed")
plt.show()
```

```python
daily_max_wind_df = daily_max_wind.reset_index()
daily_max_wind_df['Month'] = pd.to_datetime(daily_max_wind_df['Day']).dt.to_period('M')

monthly_mean_daily_max = daily_max_wind_df.groupby('Month')['Windspeed'].mean()

plt.figure(figsize=(14,5))
monthly_mean_daily_max.plot(marker='o')
plt.xlabel("Month")
plt.ylabel("Mean of Daily Max Windspeed")
plt.title("Monthly Mean of Daily Maximum Windspeed")
plt.show()
```
