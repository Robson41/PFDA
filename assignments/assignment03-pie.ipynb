```python
# ================================================================
# Assignment 03 – Pie Chart of Email Domains
# ================================================================
# This notebook loads a CSV of 1000 people, extracts the email domains,
# counts the frequency of each domain, and visualizes the results as a pie chart.
# ---------------------------------------------------------------

# Import pandas for data handling
import pandas as pd

# Import matplotlib for plotting
import matplotlib.pyplot as plt

# Import Counter (not strictly needed, but useful if we wanted custom counting)
from collections import Counter
```

```python
# ================================================================
# Cell 2 – Load CSV Data and Inspect
# ================================================================
# In this cell, we load the CSV file of 1000 people from the provided URL
# into a pandas DataFrame and check the first few rows to understand the data.

# Direct download link from Google Drive
url = "https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download"

# Load the CSV file into a pandas DataFrame
df = pd.read_csv(url)

# Display the first 5 rows to check the data
# This lets us confirm the column names and the structure of the CSV
df.head()
```

```python
# ================================================================
# Step 2: Extract email domains
# ================================================================
# Set the column name to match the CSV
email_column = 'Email'

# Extract the domain part of each email (everything after '@')
df['domain'] = df[email_column].str.split('@').str[1]

# Count how many times each domain occurs
domain_counts = df['domain'].value_counts()

# Display the top 10 domains
domain_counts.head(10)
```

```python
# ================================================================
# Step 3: Plot a pie chart of all email domains
# ================================================================

# Set figure size for clarity
plt.figure(figsize=(10, 8))

# Create the pie chart
plt.pie(
    domain_counts,                  # Values for each slice
    labels=domain_counts.index,     # Slice labels (domain names)
    autopct='%1.1f%%',              # Show percentage with 1 decimal
    startangle=140,                 # Rotate starting angle for better layout
    shadow=True,                    # Add shadow for depth
    explode=[0.05]*len(domain_counts)  # Slightly separate each slice
)

# Add a title
plt.title("Email Domain Distribution of 1000 People", fontsize=16)

# Display the chart
plt.show()
```

```python
# ================================================================
# Step 4: Pie chart for top 10 email domains
# ================================================================

# Select the top 10 most common domains
top_domains = domain_counts.head(10)

# Set figure size
plt.figure(figsize=(10, 8))

# Create the pie chart
plt.pie(
    top_domains,                  # Values for each slice
    labels=top_domains.index,     # Slice labels (domain names)
    autopct='%1.1f%%',            # Show percentage with 1 decimal
    startangle=140,               # Rotate starting angle for better layout
    shadow=True,                  # Add shadow for depth
    explode=[0.05]*len(top_domains)  # Slightly separate each slice
)

# Add a title
plt.title("Top 10 Email Domains", fontsize=16)

# Display the chart
plt.show()
```
